```python
#convert images saved in this notebook into textfiles
for i, imagefile in enumerate(ifl):
  with open('/content/drive/MyDrive/ECE1786/Project/Textfiles/'+str(pdfNames[i][:-4]), 'w') as f:
    for image in imagefile:
      f.write(pt.image_to_string(image))
```

```python
#generate csv file of just possible sentences
#take the whole line in the file where the keyword appears
import csv
with open('/content/drive/MyDrive/ECE1786/Project/Textfiles/master', 'r') as master:
  with open('/content/drive/MyDrive/ECE1786/Project/CSVs/traininglines.csv', 'w') as tl:
    csvwriter = csv.writer(tl, delimiter = ',')
    while True:
      line = master.readline()
      splitline = line.split()
      if not line:
        break
      for word in splitline:
        if word in keywords:
          csvwriter.writerow([line] + [word])
  tl.close()
master.close()




```
